```python
import folium
import requests

def search_place(query):
    base_url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": query,
        "format": "json"
    }

    response = requests.get(base_url, params=params)
    data = response.json()
    return data

# Replace 'New York City' with the place or address you want to search for
query_result = search_place("New York City")

# If there are results, get the first result's latitude and longitude
if query_result:
    first_result = query_result[0]
    lat = float(first_result['lat'])
    lon = float(first_result['lon'])
    display_name = first_result['display_name']
else:
    print("No results found.")
    exit()

# Create a folium map centered on the search result's coordinates
map_center = [lat, lon]
mymap = folium.Map(location=map_center, zoom_start=15)

# Add a marker at the search result's location
folium.Marker(location=map_center, popup=display_name).add_to(mymap)

# Display the map
mymap.save('search_result_map.html')
```

```python
import folium
import requests

def search_amenity_in_city(city_name, amenity):
    base_url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": amenity,
        "city": city_name,
        "format": "json"
    }

    response = requests.get(base_url, params=params)
    data = response.json()
    return data

# Replace 'Youssoufia' with the city name you want to search in
city_name = "Youssoufia, Marrakech-Safi, Maroc"
amenity = "masjid"

# Search for cafes in the city
cafes_result = search_amenity_in_city(city_name, amenity)

# If there are results, create a folium map centered on the city's coordinates
if cafes_result:
    city_lat = float(cafes_result[0]['lat'])
    city_lon = float(cafes_result[0]['lon'])
    map_center = [city_lat, city_lon]
    mymap = folium.Map(location=map_center, zoom_start=15)

    # Add markers for all cafes in the city
    for cafe in cafes_result:
        lat = float(cafe['lat'])
        lon = float(cafe['lon'])
        cafe_name = cafe['display_name']
        folium.Marker(location=[lat, lon], popup=cafe_name).add_to(mymap)

    # Display the map
    mymap.save('masjid_in_city_map.html')
else:
    print(f"No {amenity}s found in {city_name}.")
```

```python
def get_city_data_from_overpass(city_name):
    overpass_url = "https://overpass-api.de/api/interpreter"
    overpass_query = f"""
        [out:json];
        // gather results
        (
            // query part for the specified city
            node["place"="city"]["name"="{city_name}"];
            way["place"="city"]["name"="{city_name}"];
            relation["place"="city"]["name"="{city_name}"];
        );
        // print results
        out body;
        >;
        out skel qt;
    """
    response = requests.get(overpass_url, params={"data": overpass_query})
    data = response.json()
    return data
```
